```python

def arithmetic_arranger(problems, show_answers):
    #problems = []
    for i in problems:
      
        prob=i.split(" ")
        left=prob[0]
        l = int(left)
        operator=prob[1]
        right=prob[2]
        r=int(right)

        #make the separator
        len_l= len(left)-1
        len_r=len(right)

        if len_l >len_r:
            length_separator = len(operator + left)
            separator = "-" * length_separator
        elif len_r>=len_r:
            length_separator = len(operator + right)+1
            separator = "-" * length_separator


        #make the operator
        result = 0
        if operator == "+":
            result= l + r
        elif operator == "-":
            result= l - r
        final = str(result)
        len_f=len(final)

        #make the lines
        max_len = max(len_l,len_r)
        first_line =  " " * (max_len - len_l + 1) + left
        second_line = operator + " " + " " * (max_len - len_r) + right
        third_line=  " " * (max_len - len_f + 2) + final 

        operations = f""

        #condtions for show_answer    
        if show_answers == False:
           operations = f"{first_line}\n{second_line}\n{separator}\n"   
        elif show_answers == True:
            operations = f"{first_line}\n{second_line}\n{separator}\n{third_line}\n" 
        print(operations)
        

print(f'\n{arithmetic_arranger(["32 + 698", "3801 - 2", "45 + 43", "123 + 49","40 + 698"],False)}')
```

Conditions to respect:
- [X] There should be a single space between the operator and the longest of the two operands
- [X] The operator will be on the same line as the second operand
- [x] both operands will be in the same order as provided (the first will be the top one and the second will be the bottom).
- [] Numbers should be right-aligned.
- [] There should be four spaces between each problem.
- [#] There should be dashes at the bottom of each problem. The dashes should run along the entire length of each problem individually. (The example above shows what this should look like.)
