```python

def arithmetic_arranger(problems, show_answers=True):
    #problems = []
    for i in problems:
        prob=i.split(" ")
        left=prob[0]
        l = int(left)
        operator=prob[1]
        right=prob[2]
        r=int(right)

        #make the separator
        print(len(operator + right))

        #make the operator
        result = 0
        if operator == "+":
            result= l + r
        elif operator == "-":
            result= l - r

        #condtions for show_answer    
        if show_answers == False:
            operations = f"{left}\n{operator} {right}\n"
        elif show_answers == True:
            operations = f"{left}\n{operator} {right}\n {result}\n"

        print(operations)

print(f'\n{arithmetic_arranger(["32 + 698", "3801 - 2", "45 + 43", "123 + 49"])}')
```

Conditions to respect:
- [] There should be a single space between the operator and the longest of the two operands
- [] The operator will be on the same line as the second operand
- [] both operands will be in the same order as provided (the first will be the top one and the second will be the bottom).
- [] Numbers should be right-aligned.
- [] There should be four spaces between each problem.
- [] There should be dashes at the bottom of each problem. The dashes should run along the entire length of each problem individually. (The example above shows what this should look like.)
