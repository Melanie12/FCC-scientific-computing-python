```python
def arithmetic_arranger(problems, show_answers=False):
    #Errors
    if len(problems) > 5:
        return 'Error: Too many problems.'

    top, bottom, dash, results = [], [], [], []
    
    for prob in problems:
        
        problem = prob.split(" ")
        left = problem[0]
        operator = problem[1]
        right = problem[2]

        if not left.isdigit() or not right.isdigit():
            return "Error: Numbers must only contain digits."
        if len(left) > 4 or len(right) > 4:
            return "Error: Numbers cannot be more than four digits."
        if operator not in ('+', '-'):
            return "Error: Operator must be '+' or '-'."

        l = int(left)
        r = int(right)


        #make the operator
        result = 0
        if operator == "+":
            result = l + r
        elif operator == "-":
            result = l - r

        #make the separator
        len_l = len(left)-1
        len_r = len(right)


        width = max(len(left), len(right)) + 2
        separator = "-" * width

        final = str(result)
        len_f = len(final)

        #make the lines
        max_len = max(len_l,len_r)
        first_line  = left.rjust(width)    
        second_line = operator + " " + right.rjust(width - 2)
        separator   = "-" * width
        result_line = str(result).rjust(width)  

        top.append(first_line)
        bottom.append(second_line)
        dash.append(separator)
        results.append(result_line)  # only filled if show_answers is True

    output = '\n'.join([
        '    '.join(top),
        '    '.join(bottom),
        '    '.join(dash)])
        # include the answer line only when needed
    if show_answers:
        output += '\n' + '    '.join(results)

    return output
    
print(f'\n{arithmetic_arranger(["32 + 698", "3861 - 2", "49 + 43", "123 + 49"],True)}')
```

Conditions to respect:
- [X] There should be a single space between the operator and the longest of the two operands
- [X] The operator will be on the same line as the second operand
- [x] both operands will be in the same order as provided (the first will be the top one and the second will be the bottom).
- [X] Numbers should be right-aligned.
- [] There should be four spaces between each problem.
- [X] There should be dashes at the bottom of each problem. The dashes should run along the entire length of each problem individually. (The example above shows what this should look like.)
